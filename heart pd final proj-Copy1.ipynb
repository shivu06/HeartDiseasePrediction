```python
import pandas as ps
path="C:\\Users\\shiva\\OneDrive\\Desktop\\proj\\To predict heart disease\\framingham.csv"
data=ps.read_csv(path)

#to print the datas
print(data)
print(data.info())

print(data['RestingECG'].unique())  # to identify the different names
print(data['RestingECG'].nunique()) # to identify how many number of different names are there
data['RestingECG_n']=data['RestingECG'].map({'Normal':1,'ST':0,'LVH':2})
print(data)

import sklearn      #importing
from sklearn.preprocessing import LabelEncoder  # which is used to the encode the values 
la_sex=LabelEncoder()                        # which used to transfrom the values of specific column of company 
la_age=LabelEncoder()
la_currentSmoker=LabelEncoder()                        # which used to transfrom the values of specific column of company 
la_cigsPerDay=LabelEncoder()
la_BPMeds=LabelEncoder()
la_prevalentStroke=LabelEncoder()                        # which used to transfrom the values of specific column of company 
la_prevalentHyp=LabelEncoder()
la_diabetes=LabelEncoder()
la_totChol=LabelEncoder()
la_sysBP=LabelEncoder()
la_diaBP=LabelEncoder()
la_BMI=LabelEncoder()
la_heartRate=LabelEncoder()
la_glucose=LabelEncoder()
la_disease=LabelEncoder()

data['sex_n']=la_sex.fit_transform(data['sex'])  # these are assign to inputs in output print (company_n)new variable  and agian the old variable is assign again
data['age_n']=la_age.fit_transform(data['age'])
data['currentSmoker_n']=la_currentSmoker.fit_transform(data['currentSmoker'])
data['cigsPerDay_n']=la_cigsPerDay.fit_transform(data['cigsPerDay'])
data['BPMeds_n']=la_BPMeds.fit_transform(data['BPMeds'])
data['prevalentStroke_n']=la_prevalentStroke.fit_transform(data['prevalentStroke'])
data['prevalentHyp_n']=la_prevalentHyp.fit_transform(data['prevalentHyp'])
data['diabetes_n']=la_diabetes.fit_transform(data['diabetes'])
data['totChol_n']=la_totChol.fit_transform(data['totChol'])
data['sysBP_n']=la_sysBP.fit_transform(data['sysBP'])
data['diaBP_n']=la_diaBP.fit_transform(data['diaBP'])
data['BMI_n']=la_BMI.fit_transform(data['BMI'])
data['heartRate_n']=la_heartRate.fit_transform(data['heartRate'])
data['glucose_n']=la_glucose.fit_transform(data['glucose'])
data['disease_n']=la_disease.fit_transform(data['disease'])


inputs=data.drop(['disease_n','RestingECG','disease','sex','age','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','heartRate','glucose'],'columns')   # to seperate the rows and columns
outputs=data.drop(['disease','RestingECG_n','sex','age','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','diaBP','BMI','heartRate','glucose','RestingECG','sex_n','age_n','currentSmoker_n','cigsPerDay_n','BPMeds_n','prevalentStroke_n','prevalentHyp_n','diabetes_n','totChol_n','sysBP_n','diaBP_n','BMI_n','heartRate_n','glucose_n'],'columns')

print(inputs)
print(outputs)

#logisitic regression

import  sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.8,test_size=0.2)
#here the spilt of ip and op with train and test ,80%is used to train and 20% is used to test same order as to followed. 
print("x_train",x_train)
print("x_test",x_test)
print("y_train",y_train)
print("y_test",y_test)

# here the model is imported for the training
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train) #here the ip are used to train in model_fit
y_pred=model.predict(x_test)
print("y_pred :",y_pred)
print("y_test :",y_test)

print("LOGISITIC REGRESSION")    
from sklearn.metrics import confusion_matrix      #using confusion matrix the acc is calculated
cm=confusion_matrix(y_test,y_pred)
print("acc:",cm)


# SVM

import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.2)
from sklearn.preprocessing import StandardScaler  # increase the accuracy
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
from sklearn.svm import SVC    # importing the model
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_test)
print(y_pred)

print("SVM")
acc=model.score(x_train,y_train)*100
print("acc :",acc)




```
