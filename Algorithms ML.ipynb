# linear regression

```python
import pandas as pd   #panda is used to read files
path="C:\\Users\\shiva\\OneDrive\\Desktop\\Data\\Data\\homeprices.csv"
data=pd.read_csv(path)
print(data)          # to print the data 
print(data.info())  # guves information of storage of data 
print("no of rows",len(data))    # length of data is 
print("no of columns",len(data.columns))
print(data.size)   # size of data is shown 

inputs=data.drop('price','columns')   # to seperate the rows and columns
output=data.drop('area','columns')
print(inputs)
print(output)

import matplotlib.pyplot as plt        #to plot the graph 
plt.title("home prices prediction")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(inputs,output)
plt.grid()
plt.show()

import sklearn              # for cal of linear model
from sklearn import linear_model     # importing
model=linear_model.LinearRegression()       #initalise the model
model.fit(inputs,output)         # for traning
predict=model.predict([[3200]])        # for testing the model by giving input to test
print("predict :",predict)
acc=model.score(inputs,output)*100
print("acc :",acc)

#y=mx+c
coef=model.coef_
print("coef:",coef)
intercept=model.intercept_
print("intercept:",intercept) 
y=coef*6000+intercept #y=coef*price+intercept(slope line)
print("y:",y)
```

```python
import pandas as ps
import sklearn                #importing
from sklearn import linear_model

path="C:\\Users\\shiva\\OneDrive\\Desktop\\Data\\Data\\homeprices_Mul.csv"
data=ps.read_csv(path)


print(data)
print(data.info())

medianval = data.bedrooms.median()    # to change the value of NaN
print(medianval)

#temp= data.bedrooms.fillna(data.bedrooms.median()) #here temp value is assign using the column name 
#print(temp)

data['bedrooms']= data.bedrooms.fillna(data.bedrooms.median())

inputs = data.drop('price','columns')
output = data.drop(['area','bedrooms','age'],'columns')
print(inputs)
print(output)

model = linear_model.LinearRegression()    #intilaizing the model
model.fit(inputs,output)                   #training

res = model.predict([[3000,3,4]])    # area,3 column,bedrooms
print(res)

#y=m1x1+m2x2+m3x3+b
coef=model.coef_
print("coef:",model.coef_)
print(model.intercept_)

cal = 112.06244194*3000 +23388.88007794*4+(-3231.71790863*15)+221323.0018654
print(cal)
```

```python
import pandas as ps
path="C:\\Users\\shiva\\OneDrive\\Desktop\\Data\\Data\\Iris.csv"
data=ps.read_csv(path)

print(data)
print(data.info())

inputs=data.drop(['Id','Species'],'columns')        #output of data 
output=data.drop(['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],'columns')   #inputs to data

print("inputs:",inputs)
print("output:",output)

import sklearn 
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(inputs,output,test_size=0.2)
print("x_train :",x_train)
print("x_test:",x_test)
print("y_train:",y_train)
print("y_test:",y_test)

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=20)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print("y_pred:",y_pred)
print("y_test:",y_test)

from sklearn.metrics import confusion_matrix      #using confusion matrix the acc is calculated
cm=confusion_matrix(y_test,y_pred)
print("acc:",cm)
```

```python
import pandas as ps
path="C:\\Users\\shiva\\OneDrive\\Desktop\\Data\\Data\\Social_Network_Ads.csv"
data = ps.read_csv(path)

print(data)
print(data.info())
print(data['Gender'].unique())  # to identify the different names
print(data['Gender'].nunique()) # to identify how many number of different names are there
data['Gender_n']=data['Gender'].map({'Male':1,'Female':0})
print(data)

inputs=data.drop(['User ID','Purchased','Gender'],'columns')
outputs=data.drop(['User ID','Gender','Gender_n','Age','EstimatedSalary'],'columns')
print(inputs)
print(outputs)

import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs,outputs,train_size=0.2)

from sklearn.preprocessing import StandardScaler  # increase the accuracy
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)

from sklearn.svm import SVC    # importing the model
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_test)
print(y_pred)

acc=model.score(x_train,y_train)*100
print("acc :",acc)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print("cm=",cm)

import numpy as np
newinputs = np.array([[27,57000,0]])  #to cal predict ([[age,salary,gender_n]])
newinputs = sc.transform(newinputs)
res = model.predict(newinputs)
print(res)


```
